### **Global Business Powerhouses: A Country-Level EDA of the Forbes Global 2000**

# . Industry Profitability Comparison

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("forbes_data_cleaning.csv")
```

```python
df
```

# UNIVARIATE ANALYSIS

### 1. Categorical Columns

```python
# Example: Count of companies by Country
plt.figure(figsize=(10,6))
df['Country'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Companies')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

###  Observation
- The United States has the highest number of companies on the Forbes Global 2000 list.  
- China and Japan follow, showing their strong global business presence.  
- Most companies are concentrated in a few major economies.  
- This indicates that corporate power is mainly centered in developed countries.


```python
# Example: Count of companies by Industry
plt.figure(figsize=(12,6))
df['Industry'].value_counts().head(10).plot(kind='bar', color='orange',)
plt.title('Top 10 Industries')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.show()
```

###  Observation
- The chart shows the Top 10 industries with the most companies in the Forbes Global 2000 list.  
- Banking and Financial Services usually lead, followed by Oil & Gas, Technology, and Insurance.  
- This highlights that finance and energy sectors dominate the global economy.  
- The presence of tech firms shows the growing role of innovation-driven industries.


### 2. Numerical Columns

```python
# Sales distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Sales (INR Billion)'], kde=True, color='teal')
plt.title('Distribution of Sales')
plt.show()
```

###  Observation
- Most companies have sales in the lower to mid-range values.  
- Only a few companies reach extremely high sales, creating a right-skewed distribution.  
- This shows that a small number of large firms contribute a major share of total global sales.  
- The majority of companies operate on a smaller sales scale compared to top global giants.


```python
# Profit distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Profits (INR Billion)'], kde=True, color='purple')
plt.title('Distribution of Profits')
plt.show()
```

###  Observation
- Most companies earn moderate profits, while a few achieve very high profits.  
- The distribution is right-skewed, meaning a small number of firms dominate total profits.  
- Some companies may even have very low or negative profits, showing varied financial performance.  
- This indicates that profitability is concentrated among a limited number of top-performing companies.


```python
# Market Value distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Market Value (INR Billion)'], kde=True, color='green')
plt.title('Distribution of Market Value')
plt.show()
```

###  Observation
- Most companies have lower market values, while only a few reach extremely high valuations.  
- The distribution is right-skewed, showing that a small number of firms dominate total market capitalization.  
- This suggests that global wealth is highly concentrated among a few large corporations.  
- A majority of listed firms have moderate market sizes compared to top global giants.


# BIVARIATE ANALYSIS

### 1. Relationship between Sales & Profits

```python
industry_profit = df.groupby('Industry')['Profits (INR Billion)'].mean().sort_values(ascending=False)
industry_profit.head(10)
```

```python
plt.figure(figsize=(12,6))
industry_profit = df.groupby('Industry')['Profits (INR Billion)'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=industry_profit.values, y=industry_profit.index, palette='viridis')
plt.title('Top 10 Industries by Average Profit')
plt.xlabel('Average Profit (in millions USD)')
plt.ylabel('Industry')
plt.show()
```

###  Observation
- Banking and Financial Services industries have the highest average profits.  
- Oil & Gas, Technology, and Insurance sectors also show strong profitability.  
- These industries dominate due to large market size and consistent demand.  
- The results highlight that finance and energy remain key drivers of global profit.


```python
country_value = df.groupby('Country')['Market Value (INR Billion)'].mean().sort_values(ascending=False).head(10)
country_value
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=country_value.values, y=country_value.index, palette='mako')
plt.title('Top 10 Countries by Average Market Value')
plt.xlabel('Average Market Value (in millions USD)')
plt.ylabel('Country')
plt.show()
```

###  Observation
- The United States has the highest average market value among companies.  
- China and Japan follow, reflecting their strong corporate influence.  
- Developed economies generally host companies with higher valuations.  
- This shows that market power is concentrated in a few major global economies.


```python
plt.figure(figsize=(14,6))
industry_summary = df.groupby('Industry')[['Assets (INR Billion)', 'Market Value (INR Billion)']].mean().sort_values('Market Value (INR Billion)', ascending=False).head(10)

industry_summary.plot(kind='bar', figsize=(14,6), color=['#3498db', '#2ecc71'])
plt.title('Top 10 Industries: Average Assets vs Market Value')
plt.ylabel('Average (in millions USD)')
plt.xlabel('Industry')
plt.xticks(rotation=45)
plt.show()
```

###  Observation
- Industries like Banking, Oil & Gas, and Technology show high average assets and market values.  
- Banking firms have large asset bases, while tech firms achieve higher market value relative to assets.  
- This suggests that asset-heavy industries dominate stability, while innovation-driven sectors drive valuation.  
- Overall, asset size does not always equal higher market value â€” investor perception also matters.


# MULTIVARIATE ANALYSIS

```python
country_summary = df.groupby('Country')[['Sales (INR Billion)', 'Profits (INR Billion)', 
                                         'Assets (INR Billion)', 'Market Value (INR Billion)']].sum().sort_values('Sales (INR Billion)', ascending=False).head(10)

country_summary.plot(kind='bar', figsize=(14,7), colormap='tab20')
plt.title('Top 10 Countries by Total Financial Strength', fontsize=14)
plt.ylabel('Total (INR Billion)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend(title='Financial Metrics')
plt.show()
```

###  Observation
- The United States leads in total sales, profits, assets, and market value, showing its dominant economic power.  
- China follows closely, reflecting its strong industrial and financial growth.  
- Japan,Chaina,UK and Germany also contribute significantly to global financial strength.  
- Overall, developed nations hold most of the worldâ€™s corporate financial assets and value.


```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Sales (INR Billion)', 'Profits (INR Billion)', 'Assets (INR Billion)', 'Market Value (INR Billion)']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

###  Heatmap Shows
- The heatmap visualizes the **relationship (correlation)** between multiple numerical financial variables.  
- Each cell shows how strongly two variables move together:
  - **Positive correlation (closer to +1):** When one increases, the other also tends to increase.  
  - **Negative correlation (closer to -1):** When one increases, the other tends to decrease.  
  - **0 means no clear relationship.**
- In this dataset, most variables (Sales, Assets, Market Value) are **strongly positively correlated**, meaning big companies usually perform well across all metrics.  
- The darker or warmer colors (red) indicate a stronger relationship, while cooler colors (blue) show weaker ones.


```python
country_summary = df.groupby('Country')[['Sales (INR Billion)', 'Profits (INR Billion)', 'Assets (INR Billion)', 'Market Value (INR Billion)']].mean().sort_values('Sales (INR Billion)', ascending=False).head(10)

country_summary.plot(kind='bar', figsize=(12,6))
plt.title('Top 10 Countries by Average Financial Metrics')
plt.ylabel('Average (in billions USD)')
plt.show()
```

###  Observation
- The United States ranks highest in average sales, profits, assets, and market value per company.  
- China follow, indicating strong financial performance among their top firms.  
- Developed nations dominate in all key financial metrics, showing higher corporate efficiency and scale.  
- Emerging countries have smaller averages, reflecting differences in global economic maturity.
