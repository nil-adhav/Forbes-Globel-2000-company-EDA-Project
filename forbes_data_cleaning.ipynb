```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('A_forbes_global_1000.csv')
```

```python
df
```

```python
# Convert financial columns from millions to billions
df[["Sales", "Profits", "Assets", "Market Value"]] = df[["Sales", "Profits", "Assets", "Market Value"]] / 1000

# Optional: round for display
df[["Sales", "Profits", "Assets", "Market Value"]] = df[["Sales", "Profits", "Assets", "Market Value"]].round(2)
```

```python
df
```

```python
# Set exchange rate (1 USD ≈ 83.20 INR as of Oct 2025)
usd_to_inr = 83.20

# Convert financial columns from billions USD to billions INR
df[["Sales", "Profits", "Assets", "Market Value"]] = df[["Sales", "Profits", "Assets", "Market Value"]] * usd_to_inr

# Optional: round for display
df[["Sales", "Profits", "Assets", "Market Value"]] = df[["Sales", "Profits", "Assets", "Market Value"]].round(2)

# Optional: rename columns to reflect INR
df.rename(columns={
    "Sales": "Sales (INR Billion)",
    "Profits": "Profits (INR Billion)",
    "Assets": "Assets (INR Billion)",
    "Market Value": "Market Value (INR Billion)"
}, inplace=True)
```

```python
df
```

```python
df.to_csv('forbes_data_cleaning.csv',index=False)
```
